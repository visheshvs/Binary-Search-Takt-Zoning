# With Zoning Output

```python
import pandas as pd
import numpy as np
import datetime

# function to read CSV file and convert it into a dataframe
def read_csv_file(filename):
    df = pd.read_csv(filename, header=None)
    return df

# function to calculate the sum of all values in a dataframe
def calculate_sum(df):
    return df.sum().sum()

# function to divide the dataframe into two dataframes of equal dimensions
def divide_dataframe(df):
    row_count = df.shape[0]
    col_count = df.shape[1]
    if row_count <= col_count:
        # divide vertically from the middle
        split_index = col_count // 2
        
        #checking to make sure df1 is the one with lower sum 
        if calculate_sum(df.iloc[:, :split_index]) > calculate_sum(df.iloc[:, split_index:]):
            df1 = df.iloc[:, :split_index]
            df2 = df.iloc[:, split_index:]
        else:
            df1 = df.iloc[:, split_index:]
            df2 = df.iloc[:, :split_index]
    else:
        # divide horizontally from the middle
        split_index = row_count // 2
        if calculate_sum(df.iloc[:split_index, :]) > calculate_sum(df.iloc[split_index:, :]):
            df1 = df.iloc[:split_index, :]
            df2 = df.iloc[split_index:, :]
        else:
            df1 = df.iloc[split_index:, :]
            df2 = df.iloc[:split_index, :]

    return df1, df2

# function to perform the dataframe division recursively
def recursive_dataframe_division(df, x, flag, z, depth=0, zone=None):
    if zone is None:
        zone = pd.DataFrame(np.zeros(df.shape), index=df.index, columns=df.columns)
    if calculate_sum(df) <= x:
        #print("===============================================")
        print(f"Depth {depth} Part {flag+1} Result: ")
        print(df)
        print("Sum:",calculate_sum(df))
        print ("Zone Number:",z)
        zone[df >= 0] = z
        z = z + 1

        '''
        if flag == 1:
            zone[df >= 0] = depth
        else:
            zone[df >= 0] = depth+1
        '''
    else:
        print(f"Depth {depth} Part {flag+1}:")
        print(df)
        print("Sum:",calculate_sum(df))
        print("===============================================")
        df1, df2 = divide_dataframe(df)
        zone, z = recursive_dataframe_division(df1, x, 0, z, depth+1,zone)
        zone, z = recursive_dataframe_division(df2, x, 1, z, depth+1, zone)
    return zone,z

# main program
filename = "Classeur2.csv"
df = read_csv_file(filename)
print(df)
x = float(input("Enter the value of Target Takt: "))
print("===============================================")
flag = 0
z = 1
zone,z = recursive_dataframe_division(df, x, flag, z)
print("===============================================")
print("Resultant Zoning Configuration:")
print(zone)

# Export Zone to a CSV file with a timestamped filename
timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H-%M-%S")
zone_filename = f"Zoning {timestamp}.csv"
zone.to_csv(zone_filename, index=False, header=False)
print(f"Zone exported to {zone_filename}")
```
